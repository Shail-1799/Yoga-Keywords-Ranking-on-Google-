```python

from googleapiclient.discovery import build
import pprint        

my_api_key = "Your API Key"
my_cse_id = "Your cse ID"     # Custom Search Engine ID

def google_search(search_term, api_key, cse_id):
    service = build("customsearch", "v1", developerKey=api_key)
    res = service.cse().list(q=search_term, cx=cse_id).execute()
    return res['items']

for i,q in enumerate(Top_50_df["Keyword"]):

    results = google_search(str(q), my_api_key, my_cse_id)
    print("Result "+str(i))
    print(results)
    print("="*50)
```

# Analyzing the Response from Top 50 Keywords

```python
yoga = pd.read_csv("Top 50 Searches.csv")
```

```python
yoga = yoga.drop(yoga.columns[[0]], axis=1).iloc[:,:8] 
yoga
```

```python
yoga.to_csv("Yoga.csv")
yoga
```

```python
summary = (yoga
           .groupby(['displayLink'], as_index=False)
           .agg({'rank': ['count', 'mean']})
           .sort_values(('rank', 'count'), ascending=False)
           .assign(coverage=lambda df: df[('rank', 'count')].div(yoga['searchTerms'].nunique())))
summary.columns = ['displayLink', 'count', 'avg_rank', 'coverage']
summary['displayLink'] = summary['displayLink'].str.replace('www.', '')
summary['avg_rank'] = summary['avg_rank'].round(1)
summary['coverage'] = summary['coverage'].mul(100).round(1).astype(str).add('%')
summary.head(20).reset_index(drop=True)
```

```python
summary.iplot(kind='bar',x='displayLink',y='count',title='Domains with Highest Counts',xTitle="Display Link",yTitle='Count',color='green',theme='solar')
```

```python
top_domains = yoga['displayLink'].value_counts()[:10].index.tolist()
top_df = yoga[yoga['displayLink'].isin(top_domains)]

rank_counts = top_df.groupby(['displayLink', 'rank']).agg({'rank': ['count']}).reset_index()
rank_counts.columns = ['displayLink', 'rank', 'count']
rank_counts
```

```python
rank_counts.iplot(kind='bar',x='displayLink',y='count',title='Top Domains with Counts at their respective SERP Ranks',xTitle="Display Link",yTitle='Count at SERP Rank',color='red',theme='polar',text='rank')
```

# Analyzing Correlation with Numbers in Text


```python
import re as re
def extract_number(text):
    num = re.findall(r'[0-9]+',text)
    return (" ").join(num)
```

```python
yoga['num_title'] = yoga['title'].apply(lambda x : extract_number(x))
yoga['num_snippet'] = yoga['snippet'].apply(lambda x : extract_number(x))
```
